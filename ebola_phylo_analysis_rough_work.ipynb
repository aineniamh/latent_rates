```python
mismatches = collections.defaultdict(list)
mutation_counts = collections.defaultdict(list)
a = AlignIO.read("EBOV_Reference_Set_with_marginalAncestralStates.fasta","fasta")
aa = np.array([list(rec) for rec in a], np.character, order="F")

for node in t2.traverse("preorder"):
    if node.name=="ROOT":
        pass
    else:
        p = node.up
        print("Comparing:" + p.name + ' and ' + node.name)
        parent = [record.id for record in a if record.id==p.name][0]
        child = [record.id for record in a if record.id==node.name][0]
        print(parent, child)
        psafe = p.name.replace('|','_').replace('-','_')
        nsafe = node.name.replace('|','_').replace('-','_')
        
        aa = np.array([list(rec) for rec in a if rec.id==p.name], np.character, order="F")
        ab = np.array([list(rec) for rec in a if rec.id==node.name], np.character, order="F")
        ac = np.concatenate((aa,ab), axis=0)
        print(ac)
        p_n = psafe+"_"+nsafe
        for i in range(len(a[0])+1): #for each base up to the length of the alignment
            bp = ac[:, i:i+1] #let bp be the two bases in the pairwise alignment at that point
            if bp[0] != bp[1]: #if they're not the same, i.e. a mutation
                mutation = bp[0][0]+"_"+bp[1][0]
                mutation_counts[p_n].append(bp[0][0]+"_"+bp[1][0])
                mut_types = []

                if len(mismatches[p_n]) != 0: #if this node comparison already has a mutation
#                     print(i)
                    if mismatches[p_n][-1][1][1]==i-1: #if the last mutation added ends one base before this mutation
                        
                        #getting the mutation in the right format
                
                        combo1=zip(mismatches[p_n][-1][0].split("_"), mutation.split("_")) 
                        # x_y, p_q -split('_')-> [x, y], [p, q] -zip-> [xp,yq]
                        
                        combo2 = "".join(combo1[0])+"_"+"".join(combo1[1])
                        #[xp,yq] -join-> xp_yq
                        
                        #mutation location from previous. Location range is inclusive
                        #note this is python indexing, so actual location is one base more than i
                        location = (mismatches[p_n][-1][1][0],i) #original mutation start site, and then current base
                        
                        #mutation_type_list
                        mut_types = mismatches[p_n][-1][2]
                        
                        if mutation_dict[mutation] not in mut_types:
                            mut_types.append(mutation_dict[mutation])
                            
                        mismatches[p_n][-1]=(combo2,location,mut_types) #update the previous mutation
                    else:
                        mut_types.append(mutation_dict[mutation])
                        mismatches[p_n].append((mutation,(i,i),mut_types)) #just make a new mutation entry
                else:
                    mut_types.append(mutation_dict[mutation])
                    mismatches[p_n].append((mutation,(i,i),mut_types))#make a new mutation entry

                    
fw = open('take_2_error_profile_ebov_refs_and_internal_nodes.csv','w')
fw.write("Node_comparison,N_diffs,Type_diff,Diff_Start,Diff_End,Difference\n")
for k in mismatches:
    for j in mismatches[k]:
#         print(k, j, (len(j[0])-1)/2)
        if len(j[2])>1:
            fw.write('{},{},{};{},{},{},{}\n'.format(k,(len(j[0])-1)/2,j[2][0],j[2][1],j[1][0],j[1][1],j[0]))
        elif (len(j[0])-1)/2 > 100:
            fw.write('{},{},{},{},{},{}\n'.format(k,(len(j[0])-1)/2,j[2][0],j[1][0],j[1][1],"...".join([i[:20] for i in j[0].split("_")])))
        else:
            fw.write('{},{},{},{},{},{}\n'.format(k,(len(j[0])-1)/2,j[2][0],j[1][0],j[1][1],j[0]))
fw.close()
```
